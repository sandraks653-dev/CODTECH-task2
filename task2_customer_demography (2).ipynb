```python
import pandas as pd
import matplotlib.pyplot as plt

# Load Dataset
df = pd.read_csv("KPMG_Customer_Demography.csv")

# Basic Information
print(df.head())
print(df.info())
print(df.describe())

# Gender Distribution
gender_count = df['gender'].value_counts()

plt.figure(figsize=(6,4))
gender_count.plot(kind='bar')
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()

# Age Distribution
plt.figure(figsize=(8,5))
plt.hist(df['AGE'], bins=10)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()

# Wealth Segment
wealth = df['wealth_segment'].value_counts()

plt.figure(figsize=(6,4))
wealth.plot(kind='bar')
plt.title("Wealth Segment Distribution")
plt.xlabel("Wealth Segment")
plt.ylabel("Count")
plt.show()

# Job Industry Category
industry = df['job_industry_category'].value_counts()

plt.figure(figsize=(10,5))
industry.plot(kind='bar')
plt.title("Industry Category Distribution")
plt.xlabel("Industry")
plt.ylabel("Count")
plt.show()

# State Distribution
state = df['State'].value_counts()

plt.figure(figsize=(6,4))
state.plot(kind='bar')
plt.title("State Distribution")
plt.xlabel("State")
plt.ylabel("Count")
plt.show()

# Car Ownership
car = df['Owns_Car'].value_counts()

plt.figure(figsize=(5,5))
car.plot(kind='pie', autopct='%1.1f%%')
plt.title("Car Ownership")
plt.ylabel("")
plt.show()
```
